# Passenger Group info extraction
The passenger ID is a concat between the Group within the passenger is travelling and his number within this group.
Split this column to retrive these two information.

```python
train_data_df['Group'] = train_data_df['PassengerId'].str.split('_').str[0]
train_data_df['Group_num'] = train_data_df['PassengerId'].str.split('_').str[1]
```

Interesting info can maybe the size of the group of the passenger

```python
groupsize = train_data_df.groupby('Group').size()
train_data_df['Group_size'] = train_data_df['Group'].map(groupsize)
```

# Passenger location
we can also extract the deck, room and side from the column Cabin

```python
train_data_df['Deck'] = train_data_df['Cabin'].str.split('/').str[0]
train_data_df['Room_num'] = train_data_df['Cabin'].str.split('/').str[1]
train_data_df['Side'] = train_data_df['Cabin'].str.split('/').str[2]
```
